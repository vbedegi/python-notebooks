```python
import numpy as np

np.random.seed(42)
```

```python
from sklearn.datasets import load_iris
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import OneHotEncoder

iris = load_iris()

X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y)

encoder = OneHotEncoder(sparse=False) # !!!
encoder.fit(y.reshape(-1, 1))

y_train_ohe = encoder.transform(y_train.reshape(-1, 1))
y_test_ohe = encoder.transform(y_test.reshape(-1, 1))
```

```python
import pandas as pd

df = pd.DataFrame(X, columns=iris.feature_names)
df
```

```python
import seaborn as sns
%matplotlib inline

sns.pairplot(df, size=2)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras.callbacks import History
```

```python
model = Sequential()
model.add(Dense(16, input_shape=(4, )))
model.add(Activation('relu'))
model.add(Dropout(0.1, seed=42))
model.add(Dense(3))
model.add(Activation('softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

hist = model.fit(X_train, y_train_ohe, verbose=0, batch_size=32, epochs=400)

score = model.evaluate(X_test, y_test_ohe)

score
```

```python
sns.tsplot(data=hist.history['loss']) 
```
